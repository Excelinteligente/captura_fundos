```python
#Blibliotecas
import requests
import pandas as pd
from bs4 import BeautifulSoup

#Pega site: Rentabilidade dos Fundos de Investimento
html = requests.get('https://www37.bb.com.br/portalbb/tabelaRentabilidade/rentabilidade/gfi7,802,9085,9089,1.bbx?pk_vid=8ef034659ecbea0816686396603015d8').content

#Html "parseado"
soup = BeautifulSoup(html, 'html.parser')

#Pega todas as tabelas
tabela = soup.findAll('table',{'class':'tb_accordion table-bordered'})

#Faz laço com todas as tabelas
tab = 0
for tab in range(3):

    tabelas = str(tabela[tab])
    df = pd.read_html(tabelas)[0]

    #Muda nome das colunas
    df = df.rename(columns={'Fundos com carteira de curto prazo':'Fundo','Dia':'Rent_Dia','Acum.Mês':'Acum_Mês'})

    df = df.loc[df[df.columns[0]]!=""]
    #display(df)

    try:
        cont = 0
        for i in df:
            cont = cont + 1
            
            if df[df.columns[1]].iloc[cont]!= 'Referenciado DI' \
            and df[df.columns[1]].iloc[cont]!= 'Cambial Dólar' \
            and df[df.columns[1]].iloc[cont]!= 'Dívida Externa' \
            and df[df.columns[1]].iloc[cont]!= 'Renda Fixa' \
            and df[df.columns[1]].iloc[cont]!= 'Multimercado' \
            and df[df.columns[1]].iloc[cont]!= 'Cambial Euro':
                
                print()
                print('-' * 40)
                print(df[df.columns[0]].iloc[cont])
                print('-' * 40)

                #Formata mês atual
                new_month = str(df.columns[3])
                new_month = new_month.replace('(', '') 
                new_month = new_month.replace(')', '') 
                new_month = new_month.replace("'", '') 
                new_month = new_month.replace(',', '') 
                new_month = new_month.replace('% ', '') 

                #formata ano atual
                new_year = str(df.columns[4])
                new_year = new_year.replace('(', '') 
                new_year = new_year.replace(')', '') 
                new_year = new_year.replace("'", '') 
                new_year = new_year.replace(',', '') 
                new_year = new_year.replace('% ', '') 

                #formata 12 meses
                new_12 = str(df.columns[5])
                new_12 = new_12.replace('(', '') 
                new_12 = new_12.replace(')', '') 
                new_12 = new_12.replace("'", '') 
                new_12 = new_12.replace(',', '') 
                new_12 = new_12.replace('% ', '') 

                #formata 24 meses
                new_24 = str(df.columns[6])
                new_24 = new_24.replace('(', '') 
                new_24 = new_24.replace(')', '') 
                new_24 = new_24.replace("'", '') 
                new_24 = new_24.replace(',', '') 
                new_24 = new_24.replace('% ', '') 

                #formata 36 meses
                new_36 = str(df.columns[7])
                new_36 = new_36.replace('(', '') 
                new_36 = new_36.replace(')', '') 
                new_36 = new_36.replace("'", '') 
                new_36 = new_36.replace(',', '') 
                new_36 = new_36.replace('% ', '') 

                #Rentabilidade diária
                print('Rentabilidade diária: {}'.format(float(df[df.columns[1]].iloc[cont])/1000))

                if df[df.columns[1]].iloc[cont]!= '-':
                    print('Rentabilidade diária: {}'.format(float(df[df.columns[1]].iloc[cont])/1000))
                else:
                    print('Rentabilidade diária: -')

                if df[df.columns[2]].iloc[cont]!= '-':
                    print('Rentabilidade acumulada no mês: {}'.format(float(df[df.columns[2]].iloc[cont])/1000))
                else:
                    print('Rentabilidade acumulada no mês: -')

                if df[df.columns[3]].iloc[cont]!= '-':
                    print('{}: {}'.format(new_month, float(df[df.columns[3]].iloc[cont])/1000))
                else:
                    print('{}: {}'.format(new_month, '-')) 

                if df[df.columns[4]].iloc[cont]!= '-':
                    print('{}: {}'.format(new_year, float(df[df.columns[4]].iloc[cont])/1000))
                else:
                    print('{}: {}'.format(new_year, '-'))   

                if df[df.columns[5]].iloc[cont]!= '-':
                    print('{}: {}'.format(new_12, float(df[df.columns[5]].iloc[cont])/1000))
                else:
                    print('{}: {}'.format(new_12, '-'))        

                if df[df.columns[6]].iloc[cont]!= '-':
                    print('{}: {}'.format(new_24, float(df[df.columns[6]].iloc[cont])/1000))
                else:
                    print('{}: {}'.format(new_24, '-'))

                if df[df.columns[7]].iloc[cont]!= '-':
                    print('{}: {}'.format(new_36, float(df[df.columns[7]].iloc[cont])/1000))
                else:
                    print('{}: {}'.format(new_36, '-'))

                if df[df.columns[8]].iloc[cont]!= '-':
                    print('PL Médio: {:.3f}'.format(float(df[df.columns[8]].iloc[cont])/1000))
                else:
                    print('PL Médio: -')

                if df[df.columns[9]].iloc[cont]!= '-':
                    print('Taxa de administração: {}'.format(df[df.columns[9]].iloc[cont]))
                else:
                    print('Taxa de administração: -')
            else:
                cont = cont + 1
    except:
        print()
        #print('-' * 11 + 'Fim do procedimento' + '-' * 11)

```
